### Types of Machine Learning

Everything we've done so far has had one thing in common: we always knew the right answer for our training data. We had houses with actual sale prices. We computed loss by comparing our predictions to those known prices, and then used gradient descent to adjust our weights until the loss got as low as possible. But not every ML problem comes with answers attached. The kind of problem we've been working on is just one of three major types of machine learning, so let's take a step back to see where it fits in the bigger picture.

The type of problem we've been doing is called **supervised learning**. It's "supervised" because we're giving the model examples along with the correct answers (like a teacher supervising a student). The model looks at the input features and the known output, and learns the relationship between them. Those known outputs are called labels. In our housing example, the label for each house is its actual sale price. Supervised learning breaks down into two sub-types depending on what the label looks like. When the label is a continuous number (like a price), that's called a regression problem. When the label is a category (like "spam" or "not spam"), that's a classification problem. We've been doing regression so far, and we'll get to classification soon.

The second type is **unsupervised learning**. Here, you have data but no labels. Nobody has told you the "right answer" for any of your examples. So what can a model even do without labels? It can find patterns and structure in the data on its own. For example, say you have a dataset of customers with their purchase history, browsing behavior, and demographics. An unsupervised algorithm can look at the data and discover that there are natural clusters of customers who behave similarly. This is useful for things anomaly detection.

The third type is **reinforcement learning**. This one is pretty different from the other two, and honestly, I don't know much about it yet 😄. I don't think we'll go into it much in this series. The type of ML you use depends on what kind of data you have. If you have inputs with known correct outputs, that's supervised learning. If you have data but no labels and want to find patterns, that's unsupervised learning. Most of what we'll cover in this series is supervised learning, because it's the most common in practice and it's the foundation for understanding everything else.

One last thing. For the past couple of lessons, we've been working with one feature (square footage) and one weight because it was the simplest way to learn gradient descent and loss functions. But real problems have many features. Our original housing example had four (sqft, bedrooms, bathrooms, age), and many real world datasets have dozens or hundreds. Our `predict` function from lesson 1 calculated a prediction from multiple features by looping through lists of numbers and weights. There's a much more compact and efficient way to represent and compute with lists of numbers, and that's what we'll introduce and start using in the next lesson.

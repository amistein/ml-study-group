### Vectors and the Dot Product

In lesson 1, we wrote a predict function that looped through a list of features and a list of weights, multiplied each pair together, and summed the results.

This basic operation of mutliplying features by weights and summing them is a part of most ML models, and  is so fundamental to math, science, and engineering that it has its own name: the **dot product**. And the lists of numbers we've been passing around have a name too: **vectors**.

A vector is just a list of numbers where position matters. When we wrote [1500, 3, 2, 10] for a house's features, that was a vector. When we wrote [100, 10000, 5000, -3000] for our weights, that was also a vector. The first element means something specific (square footage), the second means something specific (bedrooms), and so on. Swapping two elements would change the meaning, which is why order matters.

The dot product of two vectors is the sum you get when you multiply corresponding elements and add them all up. So the dot product of [1500, 3, 2, 10] and [100, 10000, 5000, -3000] is 1500\*100 + 3\*10000 + 2\*5000 + 10\*(-3000) which equals 150000 + 30000 + 10000 - 30000 = 160000. If you go back and look at our predict function from lesson 1, the body of that function is literally computing a dot product and then adding a base price.

```python
# Prediction function from lesson 1
def predict(features, weights, base_price):
    return base_price + sum(f * w for f, w in zip(features, weights))
```

Why does this matter? Two reasons. First, naming the pattern lets us think about it more clearly. Instead of saying "loop through two lists, multiply pairs, sum," we can just say "take the dot product." Every time you see a dot product in machine learning (and you will see it a lot), it's doing the same thing: combining a set of values with a set of weights to produce a single number. Second, there are libraries that compute dot products (and other vector operations) much faster than a Python `for` loop. The most common one is called **NumPy**, and it's the standard Python library for numerical computation in ML. NumPy can compute a dot product of two vectors with millions of elements in a fraction of the time it would take a Python loop, because under the hood it's running optimized C code instead of interpreted Python.

NumPy's core data structure is the array, which is like a Python list but designed for fast math. You create one by passing a list to np.array():

```python
import numpy as np

features = np.array([1500, 3, 2, 10])
weights = np.array([100, 10000, 5000, -3000])
```

And you compute the dot product with np.dot():

```python
result = np.dot(features, weights)  # 160000
```

Going forward we'll be using NumPy arrays instead of plain Python lists for all of our numerical data. This is standard practice in machine learning. You'll almost never see someone train a model using plain Python lists.
